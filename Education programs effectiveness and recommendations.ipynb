```python
import pandas as pd
import numpy as np
import datetime 

#читаем исходные файлы и соединяем по ключам 
course_contents = pd.read_csv('./Data_1/Data/course_contents.csv', sep=',', encoding='utf8')


courses = pd.read_csv('./Data_1/Data/courses.csv', sep=',', encoding='utf8')

progress_phases = pd.read_csv('./Data_1/Data/progress_phases.csv', sep=',', encoding='utf8')

progresses = pd.read_csv('./Data_1/Data/progresses.csv', sep=',', encoding='utf8')

students = pd.read_csv('./Data_1/Data/students.csv', sep=',', encoding='utf8')

courses_df = progress_phases.merge(progresses, how = 'left', left_on = 'progress_id', right_on = 'id')
courses_df = courses_df.merge(course_contents, how = 'left', left_on = ['course_id','module_number','lesson_number'], right_on = ['course_id','module_number','lesson_number'])
courses_df = courses_df.merge(courses, how = 'left', left_on = 'course_id', right_on ='id')
```

```python
# рассчитываем общее количество курсов в датасете
courses_number = courses_df.groupby('title').agg('count').shape[0]

# рассчитываем  количество модулей на каждом курсе
modules_number_per_course = pd.DataFrame(courses_df.loc[:, ['title', 'module_title']].groupby(['title']).agg('nunique').loc[:,'module_title'])
modules_number_per_course

# рассчитываем  количество уроков в каждом модуле на каждом курсе
lessons_number_per_module = pd.DataFrame(courses_df.loc[:, ['title', 'module_title','lesson_title']].groupby(['title', 'module_title']).agg('nunique').loc[:,'lesson_title'])


# рассчитываем медианное количество уроков в модуле на каждом курсе
lessons_median_per_module = lessons_number_per_module.groupby(['title']).agg(['median']).loc[:, 'lesson_title']
lessons_median_per_module

# рассчитываем количество учеников на каждом курсе
students_per_course = pd.DataFrame(courses_df.loc[:, ['title', 'student_id']].groupby(['title']).agg('nunique').loc[:,'student_id'])
students_per_course

#чистим исходный датафрейм со студентами от неуказанных дат рождения и некорректно указанных дат
average_students = students.loc[:, ['id', 'birthday']].dropna()
try:
    pd.to_datetime(average_students['birthday'], format='%Y-%m-%d')
except:
    average_students['birthday'] == None
average_students.dropna(inplace = True)


#вычисляем возраст студентов в чистом датафрейме. чтобы возраст вычисления возраста были более точными, убираем экстремумы возраста- менее 11 лет и более 90
def age_calculation(date):
    today = datetime.date.today()
    date_birthday = datetime.datetime.strptime(date, '%Y-%m-%d').date() 
    age = int(round((today - date_birthday).days/365, 0))
    if age <= 11 or age >= 90:
        age = None
    else:
        age = age
    return age

average_students['age'] = average_students.loc[:, 'birthday'].apply(age_calculation)
average_students.dropna(inplace = True)


# минимальный, максимальный, средний, медианный возраст студентов
average_students_age = average_students.loc[:, 'age'].agg(['min', 'max', 'mean', 'median'])

# минимальный, максимальный, средний, медианный возраст студентов на каждом курсе
courses_df_for_age = courses_df.copy()
courses_df_for_age = courses_df_for_age.merge(average_students, how = 'inner', left_on = 'student_id', right_on ='id')
courses_df_for_age = courses_df_for_age.groupby(['title','student_id']).agg('mean')
courses_df_for_age = courses_df_for_age.groupby('title').agg(['max', 'min', 'mean', 'median']).loc[:, ['title', 'age']]
```

```python
# определяем функцию для приведения обрезания времени и выставления нужного формата даты
def to_date(date_old_1):
    date_old_1 = date_old_1[0: (date_old_1.find(' '))]
    new_date = datetime.datetime.strptime(date_old_1, '%Y-%m-%d').date()
    return new_date
# функция, которая из даты возвращает строку формата "год-месяц"
def month_calculation(date_old):
    year = str(date_old.year)
    month = str(date_old.month)
    if len(month) == 1:
        month = '0'+month
    else:
        month = month 
    month_new = str(year+"-"+month)
    return month_new

#создаем копию сводного датафрейма
courses_df_for_students_count = courses_df.copy()
#удаляем строки с пропущенными значениями
courses_df_for_students_count.dropna(inplace = True)
#создаем новую колонку с корректной датой начала урока
courses_df_for_students_count['start_date_new'] = courses_df_for_students_count['start_date'].apply(to_date)
#создаем датафрейм, отфильтрованный по признаку домашней работы и вычисляем минимальную дату начала
courses_df_for_students_count_new = pd.DataFrame(courses_df_for_students_count[courses_df_for_students_count['is_homework'] == True].groupby(['title', 'student_id']).agg('min').loc[:, 'start_date_new'])
#создаем столбец для агрегации данных по месяцв
courses_df_for_students_count_new['start_date'] = courses_df_for_students_count_new['start_date_new'].apply(month_calculation) 
```

```python
#сбрасываем индекс датафрейма для корректной группировки
courses_df_for_students_count_new.reset_index(inplace = True)
# группируем датафрейм, получая количество студентов, сделавших свое первое ДЗ на курсе в месяце
courses_df_for_students_count_final =  pd.DataFrame(courses_df_for_students_count_new.groupby(['title','start_date']).agg('count').loc[:, 'student_id'])
```

```python
# определяем фунцию для определения, закончен модуль или нет  
def module_status(date):
    if pd.isnull(date)== True:
        result = 'Not finished'
    else:
        result = 'Finished'
    return result

conversion_numerator_raw = homework_agg_cut_raw.copy()
# создаем DF, который используем в дальнейшем для вычисления конверсии: количество сдавших ДЗ в пред модуле
conversion_numerator_raw['module_status'] = conversion_numerator_raw['max_finish_date'].apply(module_status)
conversion_numerator = conversion_numerator_raw.loc[conversion_numerator_raw['module_status'] == 'Finished', ['title', 'module_number','module_status', 'student_id']].groupby(['title', 'module_number','module_status']).agg('nunique')
conversion_numerator = conversion_numerator.loc[:,['student_id']]
conversion_numerator = conversion_numerator.shift(fill_value = 0)
```

```python
# создаем DF, который используем в дальнейшем для вычисления конверсии: приступивших к ДЗ в новом модуле
conversion_denominator_raw = homework_agg_cut_raw.copy()
conversion_denominator = conversion_denominator_raw.loc[:, ['title', 'module_number', 'student_id']].groupby(['title', 'module_number']).agg('nunique')
conversion_denominator = conversion_denominator.loc[:,['student_id']]
```

```python
# содединяем два DF, вычиляем конверсию, убираем 1 модули в курсах (у них конверсии нет)
conversion_new = conversion_denominator.merge(conversion_numerator, how = 'left', left_on = ['title', 'module_number'], right_on = ['title', 'module_number'])
conversion_new_fin = conversion_new.reset_index()
conversion_new_fin = conversion_new_fin[conversion_new_fin['module_number'] != 1]
conversion_new_fin.columns = ['title', 'module_number', 'started', 'previous_finished']
conversion_new_fin['conversion'] = conversion_new_fin['started']/conversion_new_fin['previous_finished']
conversion_new_fin
```

```python
# готовим данные к визуализации на графиках, убираем лишнее
conversion_new_fin = conversion_new_fin.loc[:, ['title', 'module_number', 'conversion']]
conversion_new_fin.set_index(['title','module_number'], inplace = True)
```
